To find the critical points of this function:

$$
f(x) = (x - 1)(x - 2)(x - 3)
$$

Distribute:

$$
f(x) = x^3 - 6x^2 + 11x - 6
$$

Take the derivative $f'(x)$ and solve for $f'(x) = 0$:

$$
f'(x) = 3x^2 - 12x + 11 = 0
$$

Use the quadratic formula:

$$
\begin{equation*}
\begin{split}
x &= \frac{12 \pm \sqrt{12^2 - 4 \cdot 3 \cdot 11}}{2 \cdot 3} \\
  &= \frac{12 \pm \sqrt{12}}{6} \\
  &= \frac{12}{6} \pm \frac{\sqrt{12}}{6} \\
  &= 2 \pm \frac{\sqrt{4 \cdot 3}}{6} \\
  &= 2 \pm \frac{2\sqrt{3}}{6} \\
  &= 2 \pm \frac{\sqrt{3}}{3} \\
  &= 2 \pm 3^{\frac{1}{2}} \cdot 3^{-1} \\
  &= 2 \pm 3^{-\frac{1}{2}} \\
  &= 2 \pm \frac{1}{\sqrt{3}}
\end{split}
\end{equation*}
$$

The critical points are $2 + 1/\sqrt{3}$ and $2 - 1/\sqrt{3}$.

```python
%config InlineBackend.figure_formats = ['svg']
import math
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('../transparent.mplstyle')

sqrt3 = math.sqrt(3)
x = np.linspace(2 - sqrt3, 2 + sqrt3, 256, endpoint = True)
y = (x ** 3) - (6 * x ** 2) + (11 * x) - 6

fig = plt.figure()
ax = fig.subplots()
ax.set_xlim([x.min(), x.max()])
ax.set_ylim([y.min(), y.max()])
ax.plot(x, y, '-g', label=r'$f(x) = x^3 - 6x^2 + 11x - 6$')

xticks = [
    2 - sqrt3,
    2 - (2 / sqrt3),
    2 - (1 / sqrt3),
    2,
    2 + (1 / sqrt3),
    2 + (2 / sqrt3),
    2 + sqrt3
]

xlabels = [
    r'$2 - \sqrt{3}$',
    r'$2 - \frac{2}{\sqrt{3}}$',
    r'$2 - \frac{1}{\sqrt{3}}$',
    r'$2$',
    r'$2 + \frac{1}{\sqrt{3}}$',
    r'$2 + \frac{1}{\sqrt{3}}$',
    r'$2 + \sqrt{3}$'
]

ax.set_xticks(xticks, labels=xlabels)
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$f(x)$')
ax.legend(loc='upper left')
ax.grid(visible=True)
_ = ax.set_title(r'$x$ vs. $f(x)$')
```
